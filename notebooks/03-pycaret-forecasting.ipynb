# DATA ACQUISITION
## Import Libraries

```python
import pandas as pd
import os

from pycaret.time_series import TSForecastingExperiment
```

## Load Data

```python
# Define data paths
data_path = os.path.join('../data', 'processed')
```

```python
df = pd.read_pickle(data_path + '/grocery.pkl')
```

# Time Series

```python
# Index by Date_Received, dialy frequency
df_ts = df.set_index('Date_Received').resample('D')['Sales_Volume'].sum()
```

# Exploratory Analysis

```python
# I want to forecast the next 7 days of data and use 3-fold cross-validation
fh = 7
fold = 3
```

```python
# Intantiate Time Series Forcasting Pycaret
exp = TSForecastingExperiment()
exp.setup(data=df_ts, # data
          fh=fh, # forecast range
          target='Sales_Volume', # 
        #   log_experiment=True, # activate MLflow
        #   experiment_name='Volume_Sales_Analysis' # MLflow experiment name
          )
```

```python
# best_model, all_trained_models = exp.compare_models(include=['naive', 'grand_means', 'ets'])
best_model = exp.compare_models()
```

```python
y_predict = exp.predict_model(best_model)
y_predict
```

```python
# Plot the out-of-sample forecasts
exp.plot_model(estimator=best_model)
```

```python
# Check Goodness of Fit
exp.check_stats(best_model)
```
